# Understanding the distribution of players across different countries.

```python
fifa['nationality'].value_counts()
```

```python
fifa['nationality'].value_counts()[0:10] #Checking Top 10 countries
```

# Creating BarPlot for nationality column

```python
plt.figure(figsize=(5,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color="y")
plt.show()        
```

From the above bar plot, we can conclude that the majority of players in the dataset are from England, as it has the highest count in terms of nationality.

Creating a new DataFrame that  will contain two columns: 'Player' and 'Salary'. Each row will represent a player and their corresponding salary.

```python
player_salary = fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

# Finding the top five players who earn the highest wages.

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head()
```

# Creating a bar plot for the top five players with the highest wages

```python
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=["blue","green","red","pink","orange"])
plt.show()
```

From the above BarPlot, we can conclude that Leo Messi has the highest salary among the top five players.
